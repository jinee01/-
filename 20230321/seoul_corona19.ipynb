```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus']=False
plt.rc('font',family = 'Malgun Gothic')
```

```python
df=pd.read_csv('./data1/서울시 코로나19 확진자 현황.csv',low_memory=False)
```

```python
df.head()
```

```python
df.info()
```

```python
#데이터 확인 및 전처리  불필요한 컬럼을 제거한다. 
df=df.drop(columns=['환자번호','국적','환자정보','조치사항','이동경로','등록일','수정일','노출여부'])
```

```python
df
```

```python
#확진일을 datetime으로 변경 
df['확진일']=pd.to_datetime(df['확진일'])
df.info()
```

```python
len(df['지역'].unique())
```

```python
df['지역'].unique()
```

```python
#strip으로 중복을 제거 해야 한다.  
df['지역']=df['지역'].str.replace(' ', '' )         #str.strip() 해도 된다. 
```

```python
df
```

```python
len(df['지역'].unique())
```

```python
df['지역'].unique()
```

```python
df['지역']=df['지역'].astype('category')
```

```python
df.info()
```

```python
#결측치 확인 
df.isnull().sum()
```

```python
#구 별 확진자 동향 #확진일. 구별로 pivot_table 작성
df_gu = pd.pivot_table(df, index= '확진일',columns ='지역', values='연번',aggfunc='count',margins=True)
#df_gu = pd.pivot_table(df,'확진일','지역','연번')
df_gu
```

```python
#서울시 일별 추가확진자 동향 
s_date = df_gu['All'][ :-1]
```

```python
s_date
```

```python
#서울시 일별 추가확진자 많은 순으로 
s_date.sort_values(ascending=False)
```

```python
#서울시 일별 추가 확진자 시각화 
x= s_date.index
y = s_date.values
plt.plot(x,y)
plt.title('서울시 일별 확진자 시각화')
plt.xlabel('확진일')
plt.ylable('확진자수')
plt.xticks(rotation=45)
plt.show
```

```python
#서울시 구별 누적 확진자 비교 
s_gu = df_gu.loc['All'][:-1]
s_gu = s_gu.sort_values(ascending=False)
s_gu
```

```python
# 서울시 구별 누적확진자 많은 순 시각화 
x=s_gu.index
y=s_gu.values
plt.figure(figsize=(5,4))
plt.barh(x,y)
plt.title('서울시 구별 누적확진자수',size=10)
plt.show()
```

```python
# 서울시 최근일 누적확진자 많은 순 시각화 
s_gu1 = df_gu.iloc[-2][:-1]
s_gu1 = s_gu1.sort_values(ascending=False)
s_gu1
x=s_gu1.index
y=s_gu1.values
plt.figure(figsize=(7,6))
plt.barh(x,y)
plt.title('최근일(21-09-28 기준 지역별 확진자)',size=10)
plt.show()
```
