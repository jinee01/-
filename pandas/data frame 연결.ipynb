```python
import pandas as pd
```

```python
#pd.concat(데이터프레임 리스트, axis= 축)
df1 = pd.DataFrame([['a',1],['b',2]],columns=['letter','number'])
df2 = pd.DataFrame([['c',3],['c',4]],columns=['letter','number'])
df3 = pd.DataFrame([['e',5,'!'],['f',2,'@']],columns=['letter','number','etc'])
```

```python
print(df1)
print(df2)
print(df3)
```

```python
#컬럼명을 기준으로 연결 합치기
df_rowconcat = pd.concat([df1,df2,df3])
```

```python
df_rowconcat
```

```python
#공통된 컬럼만 연결하기   inner join
df_rowconcat2 = pd.concat([df1,df2,df3],join='inner')
df_rowconcat2
```

```python
#인덱스 재지정 
df_rowconcat3 = pd.concat([df1,df2,df3],join='inner',ignore_index=True)
df_rowconcat3
```

```python
#열로 연결하기 
df4 = pd.DataFrame({'age':[20,21,22]}, 
                    index=['amy','james','david'])
df5= pd.DataFrame({'phone':['010-1111-1111',
                            '010-2222-2222',
                            '010-3333-3333']},
                   index=['amy','james','david'])
df6= pd.DataFrame({'job':['student','programer',
                          'ceo','designer']},
                  index = ['amy','james','david','kim'])
```

```python
df_column_concat = pd.concat([df4,df5,df6])
df_column_concat
```

```python
df_column_concat = pd.concat([df4,df5,df6],axis=1)
df_column_concat
```

```python
df_column_concat2 = pd.concat([df4,df5,df6],axis=1,join ='inner')
df_column_concat2
```
