```python
#누락값(NaN)을 앞데이터로 채우기 
df = df.fillna(method ='ffill')    #forward로 채워라 ffill  
df                                 # 데이터를 보고 어떤 값으로 채울지 정함 
```

```python
#서울에서 다른 지역으로 이동한 데이터만 추출하여 정리 
mask = (df['전출지별'] =='서울특별시') & (df['전입지별'] != '서울특별시') 
mask
```

```python
df_seoul = df[mask]
df_seoul
```

```python
df_seoul = df_seoul.drop(['전출지별'], axis = 1)
df_seoul
```

```python
df_seoul.rename({'전입지별':'전입지'}, axis = 1, inplace = True)
df_seoul
```

```python
df_seoul.set_index('전입지', inplace = True)
df_seoul.head(3)
```

```python
#서울에서 경기도로 이동한 인구 데치터 값만 선택 
sr_one = df_seoul.loc['경기도']
```

```python
plt.plot(sr_one.index , sr_one.values)
```

```python
plt.plot(sr_one)
```

```python
plt.plot(sr_one)
#차트의 제목을 추가하는 방법 
plt.title('서울 -> 경기 인구 이동')
#축의 이름 
plt.xlabel('기간')
plt.ylabel('인구 이동수')
plt.show()
```

```python
#주피터에 한글이 없기 때문에 폰트를 지정해주는 코드 
 #ttf 는 윈도우 안의 폰트를 활용한다는 의미
    #따로 설치할 필요 없고 matplot에서 불러오면 된다. 
```

```python
plt.figure(figsize=(12,6))                                                #그림사이즈 세로 인치 
plt.plot(sr_one.index, sr_one.values, marker='o', markersize=7)
plt.title('서울 -> 경기 인구 이동',size=22)
plt.xlabel( '기간',size=18)
plt.xticks(size= 10, rotation=45)                                         #사이즈 #각도조절 
plt.ylabel('인구 이동수',size=18)
plt.ylim(5000,700000)                                                     #limit를 늘릴 수 있다. (최솟값, 최대값)
plt.legend(fontsize = 15, labels=['서울 ->경기'],loc = 'upper left')      #범례 
plt.show() #변경사항을 저장하고 그래프로 출력 
```

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(sr_one, marker='o', markersize=7)
ax2.plot(sr_one, marker='v', markerfacecolor='olive',
         markersize=7, color='olive', linewidth=4, 
         label ='서울 ->경기')

ax2.legend(loc='best')

ax1.set_ylim(50000,700000 )
ax2.set_ylim(50000,700000)

ax1.set_xticklabels(sr_one.index, rotation=75)
ax2.set_xticklabels(sr_one.index, rotation=75)   

plt.show() 
```

```python
#스타일 서식 지정 

plt.style.use('ggplot')
fig = plt.figure(figsize= (20,5))
ax = fig.add_subplot(1,1,1)
ax.plot(col_years, df3.loc['충청남도',:],marker= 'o', markerfacecolor='green',             
       markersize=10,color='olive',linewidth=2,label='서울->충남')          #[행, 열전체]

plt.style.use('ggplot')
fig = plt.figure(figsize= (20,5))

ax.plot(col_years, df3.loc['경상북도',:],marker= 'o', markerfacecolor='skyblue',             
       markersize=10,color='blue',linewidth=2,label='서울->경북') 

plt.style.use('ggplot')
fig = plt.figure(figsize= (20,5))

ax.plot(col_years, df3.loc['강원도',:],marker= 'o', markerfacecolor='red',             
       markersize=10,color='magenta',linewidth=2,label='서울->강원') 
ax.legend(loc='best')
ax.set_xlabel('기간',size =12)
ax.set_ylabel('인구수',size =12)
ax.set_xticklabels(col_years, rotation=90)

plt.show()
```

```python
df4= df_seoul.loc[['충청남도','경상북도','강원도','전라남도'], col_years]

plt.style.use('ggplot')
fig = plt.figure(figsize= (20,5))
ax1 = fig.add_subplot(2,2,1)
ax.plot(col_years, df4.loc['충청남도',:],marker= 'o', markerfacecolor='green',             
       markersize=10,color='green',linewidth=2,label='서울->충남')    
ax1.legend(loc='best')
ax1.set_xlabel('기간',size =12)
ax1.set_ylabel('인구수',size =12)
ax1.set_xticklabels(col_years, rotation=90)

plt.style.use('ggplot')
fig = plt.figure(figsize= (20,5))
ax2 = fig.add_subplot(2,2,2)
ax2.plot(col_years, df4.loc['경상북도',:],marker= 'o', markerfacecolor='skyblue',             
       markersize=10,color='skyblue',linewidth=2,label='서울->경북') 
ax2.legend(loc='best')
ax2.set_xlabel('기간',size =12)
ax2.set_ylabel('인구수',size =12)
ax2.set_xticklabels(col_years, rotation=90)

plt.style.use('ggplot')
fig = plt.figure(figsize= (20,5))
ax3 = fig.add_subplot(2,2,3)
ax3.plot(col_years, df4.loc['강원도',:],marker= 'o', markerfacecolor='red',             
       markersize=10,color='red',linewidth=2,label='서울->강원') 
ax3.legend(loc='best')
ax3.set_xlabel('기간',size =12)
ax3.set_ylabel('인구수',size =12)
ax3.set_xticklabels(col_years, rotation=90)

plt.style.use('ggplot')
fig = plt.figure(figsize= (20,5))
ax4 = fig.add_subplot(2,2,4)
ax4.plot(col_years, df4.loc['전라남도',:],marker= 'o', markerfacecolor='yellow',             
       markersize=10,color='yellow',linewidth=2,label='서울->전남') 

ax4.legend(loc='best')
ax4.set_xlabel('기간',size =12)
ax4.set_ylabel('인구수',size =12)
ax4.set_xticklabels(col_years, rotation=90)

plt.show()
```
