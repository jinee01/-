```python
import pandas as pd
#melt를 통해 데이터 재구조화
df = pd.DataFrame({"item": ["shirts", "shirts", "shirts", "shirts", "shirts",
                          "pants", "pants", "pants", "pants"],
                    "color": ["white", "white", "white", "black", "black",
                          "white", "white", "black", "black"],
                   "size": ["small", "large", "large", "small",
                          "small", "large", "small", "small",
                         "large"],
                   "sale": [1, 2, 2, 3, 3, 4, 5, 6, 7],
                   "inventory": [2, 4, 5, 5, 6, 6, 8, 9, 9]})
```

```python
df
```

```python
#집계분석 item과 size별 재고 합계를 pivot_table 메소드로 분석
#재고,pants와 shirt large small
df.pivot_table(index=['item','color'],columns='size',
               values='inventory',aggfunc='sum',fill_value=0)
```

```python
df.pivot_table(index=['item','color'],columns='size',
               values=['inventory','sale'],aggfunc='sum',fill_value=0)
```

```python
import seaborn as sns 
df_titanic = sns.load_dataset('titanic')
df_titanic =df_titanic[['survived','pclass','sex','age','embarked']]
```

```python
df_titanic.dropna(inplace=True)
df_titanic.head()
```

```python
#성별 객실 등급별 승선지수 count를 사용 
#df.pivot_table(index=['sex',],columns='pclass',
 #              values=[1,2,3],aggfunc='sum',fill_value=0)
```

```python
df_titanic.pivot_table(index='sex',columns='pclass',
                      values='survived', aggfunc='count',
                      margins=True)
#survive = [1,0,1,1,0] 5명
```

```python
df_titanic.pivot_table(index='sex',columns='pclass',
                      values='age', aggfunc='count',
                      margins=True)
```

```python
#성별 객실 등급별 생존자수
df_titanic.pivot_table(index='sex',columns='pclass',
                      values='survived', aggfunc='sum',
                      margins=True)
#survive = 1+0+1+1+0
```

```python
#성별 객실 등급별 생존율 *********** aggfunction의 mean은 default 임 ***********
df_titanic.pivot_table(index='sex',columns='pclass',
                      values='survived', aggfunc='mean',
                      margins=True)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.info()
```

```python
tips['tip_pct'] = ((tips.tip/tips.total_bill)*100).round(2)
```

```python
tips.pivot_table('tip_pct','sex')
```

```python
tips.pivot_table('tip_pct','sex','smoker',aggfunc='count',
                      margins=True,margins_name='계')
```

```python
tips.pivot_table('tip_pct',['sex','smoker'])
```

```python
tips.pivot_table('tip_pct','sex','smoker')
```

```python
tips.pivot_table('tip_pct','smoker',aggfunc=['mean','min','max'])   #shift+tab 눌러서 설명 보기 
```

```python
#df.groupby(그룹기준컬럼).통계적용컬럼.통계함수
#.count(): 누락값을 제외한 데이터 수 
#.size(): 누락값을 포함한 데이터 수
```

```python
df_titanic1 = df_titanic.groupby('pclass').survived.count().to_frame()
df_titanic1
```

```python
#객실등급 별 승선자 수 결과 집계
df_titanic2 = df_titanic.groupby('pclass').survived.sum().to_frame()
df_titanic2
```

```python
df_titanic3 = df_titanic.groupby('pclass').survived.mean().to_frame()
df_titanic3
```

```python

df_titanic4 =pd.concat([df_titanic1,df_titanic2,df_titanic3],axis=1)
df_titanic4.columns =['승선자수','생존자수','생존비율']
df_titanic4
```

```python
df_titanic5 = df_titanic.groupby('sex').survived.count().to_frame()
df_titanic5
```

```python
df_titanic6 = df_titanic.groupby('sex').survived.sum().to_frame()
df_titanic6
```

```python
df_titanic7 = df_titanic.groupby('sex').survived.mean().to_frame()
df_titanic7
```

```python
df_titanic8 =pd.concat([df_titanic5,df_titanic6,df_titanic7],axis=1)
df_titanic8.columns =['승선자수','생존자수','생존비율']
df_titanic8
```
