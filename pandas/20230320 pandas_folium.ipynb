```python
# 대학교 리스트를 데이터 프레임으로 변환
df = pd.read_excel('./data1/서울지역 대학교 위치.xlsx',engine='openpyxl')
df = df.reset_index()
df = df.set_index(['Unnamed: 0'])
df.index
```

```python
#서울 지도를 먼저 만들고 지리적 위치를 표시한다. 
seoul_map = folium.Map(location=[37.55,126.98],tiles = 'Stamen Terrain', zoom_start=12)
```

```python
for name, lat, lng in zip(df.index, df.위도 , df.경도):
        folium.Marker([lat,lng],popup=name).add_to(seoul_map)
```

```python
#경기도 인구변화 데이터를 불러와서 데이터 프레임으로 변환
df2= pd.read_excel('./경기도인구데이터.xlsx', index_col='구분',engine='openpyxl')
df2.columns = df2.columns.map(str)
```

```python
#경기도 시군구 경계 정보를 가진 geo-json 파일 불러오기  try except 문으로 불러온다. 
geo_path='./경기도행정구역경계.json'
try:
    geo_data = json.load(open(geo_path, encoding='utf-8'))
except:
    geo_data = json.load(open(geo_path, encoding='utf-8-sig'))
```

```python
#경기도 지도만들기 
g_map = folium.Map(location=[37.5502,126.982], titles='Stamen Terrian',zoom_start = 9)
```

```python
#출력할 연도를 선택(2007~2017)   #최대최소 먼저 확인하고 threshold_scale을 결정해야 한다. 
year = '2017'
#단계구분도(Chorpleth) 단계구분도 표시하기
folium.Choropleth(geo_data=geo_data , data=df2[year],
                columns=[df2.index, df2[year]],
                fill_color='YlOrRd', fill_opacit=0.7,
                line_opacity=0.3,
                threshold_scale=[10000, 100000, 300000, 500000, 700000],
                key_on ='feature.properties.name').add_to(g_map)
g_map
```

```python
g_map.save('./gyonggi_population'+year +'.html')
```
